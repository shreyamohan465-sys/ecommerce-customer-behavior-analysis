```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("ecommerce_customer_behavior_dataset.csv")

data.head()
```

```python
data = pd.read_csv("ecommerce_customer_behavior_dataset_v2.csv")

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data['Total_Amount'].sum()
```

```python
data['Total_Amount'].mean()
```

```python
data.groupby('Product_Category')['Total_Amount'].sum()
```

```python
data.groupby('Product_Category')['Total_Amount'].sum().plot(kind='bar', figsize=(8,5))

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='Gender', data=data)

plt.title("Customer Gender Distribution")
plt.show()
```

```python
data.groupby('Payment_Method')['Total_Amount'].sum()
```

```python
data.groupby('Payment_Method')['Total_Amount'].sum().plot(kind='pie', autopct='%1.1f%%')

plt.title("Sales by Payment Method")
plt.ylabel("")
plt.show()
```

```python
sns.histplot(data['Age'], bins=20)

plt.title("Customer Age Distribution")
plt.show()
```

```python
data.groupby('City')['Total_Amount'].sum().sort_values(ascending=False).head(10)
```

```python
data.groupby('City')['Total_Amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Top Cities by Sales")
plt.show()
```

```python
data['Is_Returning_Customer'].value_counts()
```

```python
sns.countplot(x='Is_Returning_Customer', data=data)

plt.title("Returning vs New Customers")
plt.show()
```

```python
data.to_csv("cleaned_ecommerce_data.csv", index=False)
```
